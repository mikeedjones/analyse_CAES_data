Electrons are placed at various z positions in the extaction region ,in simulation, and accelerated towards the MCP. The time the electrons take to reach the MCP form each z position is recorded for various extraction voltages.

```python
import fit_TOF_data as TOF
import matplotlib.pyplot as plt
import numpy as np
```

The time of flight for electrons with different z positions at each extraction voltage.

```python
TOF.load_TOF_model('TOF_scan/1k')
TOF.show_TOF_model('TOF_scan/1k')
```

Both the ns blue laser pulse and the pluse from the electron bunch hitting the MCP were recoreded on an oscilliscope scope. Gaussain functions were fitted to the two pulses and the time between them, $t_r$, was recorded. 

$t_r$ where used to fit
$$t_r=t_0(U,z)+\Delta t,$$ where U is the extraction voltage, $z$ is the z position of the source, $\Delta t$ is the sum of the signal line delays of the two signals, and $t_0(U,z)$ is the time of flight calculated by the model.

The z position of the MOT was measured by the MOT cameras as 0.068 mm. Using that z position the TOF plots below were produced for a range of different einzel/correction lens biases.

```python
plt.clf()
plt.subplots(nrows=2, ncols=3, figsize=(13, 10), sharey=True)
plt.subplot(2,2,1)
TOF.TOF_versus_extV('../../experimental_data/17-04-18/TOF_1',TOF_mod_dir='TOF_scan/1k',disable_prog_bar=True)
plt.title('1 kV on correction lens and einzel lens')
plt.subplot(2,2,2)
TOF.TOF_versus_extV('../../experimental_data/17-04-18/TOF_2',TOF_mod_dir='TOF_scan/1_5k',disable_prog_bar=True)
plt.title('1.5 kV on correction lens and einzel lens')
plt.subplot(2,2,3)
TOF.TOF_versus_extV('../../experimental_data/17-04-18/TOF_3',TOF_mod_dir='TOF_scan/2k',disable_prog_bar=True)
plt.title('2 kV on correction lens and einzel lens')
plt.subplot(2,2,4)
TOF.TOF_versus_extV('../../experimental_data/17-04-18/TOF_4',TOF_mod_dir='TOF_scan/2_5k',disable_prog_bar=True)
plt.title('2.5 kV on correction lens and einzel lens')
plt.show()
plt.clf()
```

The data fits the model well and gives a $\Delta$t of $-3.15 \pm 0.04 $ ns. The variation in $\Delta t$ between the measurements can be attributed to path length differences - the addional 0.11 ns would only require the path to be ~3 mm longer for the maximum extraction voltage.
